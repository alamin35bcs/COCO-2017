```python
import matplotlib.pyplot as plt
from IPython.display import display, Image
import os

# Use journal-style font
plt.rcParams['font.family'] = 'serif'

# ---------------------------
# Figure 1(A): Publications
# ---------------------------
years = list(range(2016, 2027))
research = [150, 220, 300, 380, 450, 540, 630, 720, 800, 320, 100]
review =   [5, 10, 15, 20, 30, 80, 90, 100, 110, 40, 10]

x = list(range(len(years)))

plt.figure(figsize=(8,5))
plt.bar(x, research, width=0.4, label='Research Articles')
plt.bar([i + 0.4 for i in x], review, width=0.4, label='Review Articles')

plt.xticks([i + 0.2 for i in x], years)
plt.xlabel("Year")
plt.ylabel("Number of Publications")
plt.title("Figure 1(A): Publications per Year")
plt.legend()
plt.tight_layout()

plt.savefig("Figure_1A.pdf", dpi=600, bbox_inches='tight')
plt.savefig("Figure_1A.png", dpi=600, bbox_inches='tight')
plt.show()   # This displays the figure in Colab
plt.close()

# ---------------------------
# Figure 1(B): Applications
# ---------------------------
labels = ['Biomedical', 'Aerospace', 'Automotive', 'Energy & WAAM', 'Other']
sizes = [24, 21, 18, 17, 20]

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.0f%%', startangle=90)
plt.title("Figure 1(B): Application Distribution (N=56)")
plt.tight_layout()

plt.savefig("Figure_1B.pdf", dpi=600, bbox_inches='tight')
plt.savefig("Figure_1B.png", dpi=600, bbox_inches='tight')
plt.show()   # This displays the figure in Colab
plt.close()

# ---------------------------
# Check saved files
# ---------------------------
print("Saved files:")
print(os.listdir())
```
