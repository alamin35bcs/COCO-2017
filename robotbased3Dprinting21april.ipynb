```python
import matplotlib.pyplot as plt

subject_areas = [
    "Engineering",
    "Computer Science",
    "Materials Science",
    "Medicine",
    "Physics and Astronomy",
    "Biochemistry, Genetics and Molecular Biology",
    "Mathematics",
    "Chemical Engineering",
    "Business, Management and Accounting",
    "Chemistry"
]

subject_counts = [322, 149, 114, 77, 64, 53, 41, 35, 34, 33]

plt.figure(figsize=(10, 6))
plt.bar(subject_areas, subject_counts)

plt.xlabel("Subject Area", fontsize=12)
plt.ylabel("Number of Documents", fontsize=12)
plt.title("Top 10 Subject Areas in Scopus Search Results", fontsize=14)

plt.xticks(rotation=45, ha="right")

for i, value in enumerate(subject_counts):
    plt.text(i, value + 3, str(value), ha='center', fontsize=9)

plt.tight_layout()
plt.savefig("top_10_subject_area_bar_chart.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import os

# ==============================
# Application / Topic Distribution
# Based on 56 selected papers
# ==============================

labels = [
    "Biomedical",
    "Automotive",
    "Construction / Mobile Robot",
    "Aerospace",
    "Energy & WAAM",
    "AI / Sensors / Quality Monitoring",
    "General / Materials / Systems"
]

counts = [8, 12, 8, 3, 3, 8, 14]

# ==============================
# Save path: Downloads/Graph folder
# ==============================

graph_folder = os.path.join(os.path.expanduser("~"), "Downloads", "Graph")
os.makedirs(graph_folder, exist_ok=True)

save_path = os.path.join(graph_folder, "application_area_distribution_56_papers.png")

# ==============================
# Pie Chart
# ==============================

plt.figure(figsize=(10, 7))

plt.pie(
    counts,
    labels=labels,
    autopct="%1.1f%%",
    startangle=90,
    pctdistance=0.70,
    labeldistance=1.10,
    textprops={"fontsize": 11, "fontname": "Times New Roman"}
)

plt.title(
    "Distribution of Application and Technical Areas Based on 56 Selected Papers",
    fontsize=14,
    fontname="Times New Roman"
)

plt.tight_layout()
plt.savefig(save_path, dpi=300, bbox_inches="tight")
plt.show()

print("Figure saved to:", save_path)
print("Total papers:", sum(counts))
```

```python
import matplotlib.pyplot as plt

application_areas = [
    "Biomedical",
    "Aerospace",
    "Automotive",
    "Energy & WAAM",
    "Construction"
]

paper_counts = [13, 12, 10, 10, 11]

plt.figure(figsize=(9, 7))

wedges, texts, autotexts = plt.pie(
    paper_counts,
    labels=application_areas,
    autopct='%1.0f%%',
    startangle=90,
    pctdistance=0.65,
    labeldistance=1.12,
    textprops={'fontsize': 12, 'fontname': 'Times New Roman'}
)

for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_fontname("Times New Roman")

plt.title(
    "Distribution of Application Areas Based on the 56 Selected Papers",
    fontsize=14,
    fontname="Times New Roman"
)

plt.tight_layout()
plt.savefig("application_area_pie_chart_journal.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# ==============================
# Scopus Publications by Year
# ==============================

years = np.array([2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025, 2026])
publications = np.array([5, 4, 8, 23, 42, 45, 62, 59, 105, 123, 56])

# ==============================
# Save path: Downloads/Graph folder
# ==============================

graph_folder = os.path.join(os.path.expanduser("~"), "Downloads", "Graph")
os.makedirs(graph_folder, exist_ok=True)

save_path = os.path.join(graph_folder, "scopus_publications_per_year.png")

# ==============================
# Bar Chart
# ==============================

plt.figure(figsize=(10, 5.5))

plt.bar(years, publications, width=0.65)

plt.xlabel("Year", fontsize=12, fontname="Times New Roman")
plt.ylabel("Number of Publications", fontsize=12, fontname="Times New Roman")
plt.title("Publications per Year on Robot-Based 3D Printing (2016–2026)",
          fontsize=14, fontname="Times New Roman")

plt.xticks(years, fontsize=11, fontname="Times New Roman")
plt.yticks(fontsize=11, fontname="Times New Roman")

for x, y in zip(years, publications):
    plt.text(x, y + 2, str(y), ha="center", va="bottom",
             fontsize=10, fontname="Times New Roman")

plt.tight_layout()
plt.savefig(save_path, dpi=300, bbox_inches="tight")
plt.show()

print("Figure saved to:", save_path)
```
